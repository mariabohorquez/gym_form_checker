# FASE 2: Fine-Tuning de YOLOv8 para Clasificación
Vamos a entrenar YOLOv8n, YOLOv8s y YOLO11n en el dataset.

```
import os
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import Image, display
import shutil

print('CUDA disponible:', torch.cuda.is_available())
print('GPU:', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'None')
```

## 1. Entrenamiento de Modelos

```
dataset_path = os.path.abspath('Exercise-Frame-Labelling-8')

def train_model(model_name, epochs=50, imgsz=224):
    print(f'\nEntrenando {model_name}...')
    model = YOLO(f'{model_name}-cls.pt')
    results = model.train(
        data=dataset_path,
        epochs=epochs,
        imgsz=imgsz,
        device=0 if torch.cuda.is_available() else 'cpu',
        project='runs/cls',
        name=model_name,
        exist_ok=True
    )
    return results

# Entrenar YOLOv8n-cls
results_v8n = train_model('yolov8n')
# Entrenar YOLOv8s-cls
results_v8s = train_model('yolov8s')
# Entrenar YOLO11n-cls
results_v11n = train_model('yolo11n')
```

## 2. Evaluación de Resultados
Visualicemos la matriz de confusión y curvas de entrenamiento del mejor modelo esperado (YOLOv8s).

```
best_model_dir = 'runs/cls/yolov8s'

# Mostrar Matriz de confusión
cm_path = os.path.join(best_model_dir, 'confusion_matrix.png')
if os.path.exists(cm_path):
    display(Image(filename=cm_path))
    shutil.copy(cm_path, '../data/yolov8s_confusion_matrix.png')

# Mostrar Curvas de entrenamiento
results_path = os.path.join(best_model_dir, 'results.png')
if os.path.exists(results_path):
    display(Image(filename=results_path))
    shutil.copy(results_path, '../data/yolov8s_training_curves.png')
```
