# FASE 1: EDA y Descarga del Dataset
En este notebook nos encargaremos de descargar el dataset de Roboflow, analizar la distribución de clases, visualizar algunas muestras, analizar los splits (train/val/test) y la distribución de aspect ratios.

```python
import os
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import cv2
from roboflow import Roboflow
from pathlib import Path

load_dotenv()
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Descarga del Dataset desde Roboflow

```python
api_key = os.getenv('ROBOFLOW_API_KEY')
rf = Roboflow(api_key=api_key)
project = rf.workspace('yolov8-training-ipgoa').project('exercise-frame-labelling')
dataset = project.version(8).download('folder')
print(f'Dataset descargado en: {dataset.location}')
```

## 2. Análisis de Distribución de Clases y Splits

```python
dataset_path = Path(dataset.location)
splits = ['train', 'valid', 'test']
classes = [d for d in os.listdir(dataset_path / 'train') if os.path.isdir(dataset_path / 'train' / d)]

data_counts = {split: {cls: 0 for cls in classes} for split in splits}
for split in splits:
    for cls in classes:
        cls_path = dataset_path / split / cls
        if cls_path.exists():
            data_counts[split][cls] = len(list(cls_path.glob('*.jpg')))

import pandas as pd
df_counts = pd.DataFrame(data_counts)
df_counts['total'] = df_counts.sum(axis=1)
display(df_counts)

plt.figure(figsize=(10, 6))
sns.barplot(x=df_counts.index, y=df_counts['total'], palette='viridis')
plt.title('Distribución Total de Clases')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Imágenes')
plt.savefig('../data/class_distribution.png')
plt.show()

df_counts_splits = df_counts.drop(columns=['total'])
df_counts_splits.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Distribución de Clases por Split')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Imágenes')
plt.legend(title='Split')
plt.savefig('../data/split_distribution.png')
plt.show()
```

## 3. Visualización de Muestras

```python
fig, axes = plt.subplots(len(classes), 5, figsize=(15, 3 * len(classes)))

for i, cls in enumerate(classes):
    cls_path = dataset_path / 'train' / cls
    images = list(cls_path.glob('*.jpg'))
    samples = np.random.choice(images, 5, replace=False) if len(images) >= 5 else images
    
    for j, img_path in enumerate(samples):
        img = cv2.imread(str(img_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axes[i, j].imshow(img)
        axes[i, j].axis('off')
        if j == 0:
            axes[i, j].set_title(cls, fontsize=14, loc='left')

plt.tight_layout()
plt.savefig('../data/sample_images.png')
plt.show()
```

## 4. Distribución de Aspect Ratios

```python
aspect_ratios = []
widths = []
heights = []
all_train_images = list((dataset_path / 'train').rglob('*.jpg'))
sample_images = np.random.choice(all_train_images, min(1000, len(all_train_images)), replace=False)

for img_path in sample_images:
    img = cv2.imread(str(img_path))
    if img is not None:
        h, w, _ = img.shape
        widths.append(w)
        heights.append(h)
        aspect_ratios.append(w / h)

plt.figure(figsize=(10, 6))
sns.histplot(aspect_ratios, bins=30, kde=True, color='purple')
plt.title('Distribución de Aspect Ratios (Ancho / Alto)')
plt.xlabel('Aspect Ratio')
plt.ylabel('Frecuencia')
plt.savefig('../data/aspect_ratios.png')
plt.show()

print(f'Dimensiones promedio: {np.mean(widths):.0f}x{np.mean(heights):.0f}')
```
